# Analysis of the scores for 5 classifiers

The classifiers are:

* **Static**: static features, 188 dimensions
* **Dynamic**: dynamic features 16 dimensions
* **Combined**: combined features 204 dimensions
* **Choser**: Logical operations over static and dynamic features, *or, and, only_static, only_dynamic*
* **MKL**: Multi-Kernels approach with a mixing factor between 0 and 1 using static and dynamic features

All the classifiers are aggressively **bred for generalization**: Cross validation is performed by generating 100 sequences with 3 objects left for testing. The results below are reported an additional dataset that the classifiers have never seen before.

```python
import sklearn
import load_static_features
import utilities
from collections import defaultdict
import cPickle
import os
import train_combined_features
from sklearn.metrics import f1_score
from best_of_them import BestOfThem
import train_adjectives_features
%pylab inline
#figsize(15,5)
figsize(10,30)
```

#### Loading Static Classifiers

```python
path = "/home/pezzotto/log/static_features/trained_adjectives"
static_classifiers = {}
for adj in utilities.adjectives:
    filename = "trained_" + adj + ".pkl"
    f = cPickle.load(open(os.path.join(path, filename))) 
    static_classifiers[adj]= f
```

#### Loading Combined Classifiers

```python
path = "/home/pezzotto/log/combined_features/trained_adjectives"
combined_classifiers = {}
for adj in utilities.adjectives:
    filename = "trained_" + adj + ".pkl"
    f = cPickle.load(open(os.path.join(path, filename))) 
    combined_classifiers[adj]= f
```

#### Loading Dynamic Classifiers

```python
path = "/home/pezzotto/log/hmm_features/trained_adjectives"
dynamic_classifiers = {}
for adj in utilities.adjectives:
    filename = "trained_" + adj + ".pkl"
    f = cPickle.load(open(os.path.join(path, filename))) 
    dynamic_classifiers[adj]= f
```

#### Loading Chooser Classifiers

```python
path = "/home/pezzotto/log/best_of_them/trained_adjectives"
chooser_classifiers = {}
for adj in utilities.adjectives:
    filename = "trained_" + adj + ".pkl"
    f = cPickle.load(open(os.path.join(path, filename))) 
    chooser_classifiers[adj]= f
```

#### Loading Data Set

```python
static_train_test_set = train_adjectives_features.load_adjective_collective("/home/pezzotto/log/static_features")
dynamic_train_test_set = train_adjectives_features.load_adjective_collective("/home/pezzotto/log/hmm_features")
combined_train_test_set = cPickle.load(open("/home/pezzotto/log/combined_features/combined_dict.pkl"))
```

#### Collecting Scores

```python
static_scores = {}
static_scores_list = []
combined_scores = {}
combined_scores_list = []
dynamic_scores = {}
dynamic_scores_list = []
chooser_scores = {}
chooser_scores_list = []
mkl_scores = utilities.mkl_scores
mkl_scores_list = []
for adj in utilities.adjectives:
    #static
    test_X = static_train_test_set[adj]['test']['X']
    test_Y = static_train_test_set[adj]['test']['Y']
    clf = static_classifiers[adj]['classifier']
    clf_out = clf.predict(test_X)
    v = f1_score(test_Y, clf_out)
    static_scores[adj] = v
    static_scores_list.append(v)
    
    #dynamic
    test_X = dynamic_train_test_set[adj]['test']['X']
    test_Y = dynamic_train_test_set[adj]['test']['Y']
    clf = dynamic_classifiers[adj]['classifier']
    clf_out = clf.predict(test_X)
    v = f1_score(test_Y, clf_out)
    dynamic_scores[adj] = f1_score(test_Y, clf_out)
    dynamic_scores_list.append(v)
    
    #combined
    test_X = combined_train_test_set[adj]['test']['X']
    test_Y = combined_train_test_set[adj]['test']['Y']
    clf = combined_classifiers[adj]['classifier']
    clf_out = clf.predict(test_X)
    v = f1_score(test_Y, clf_out)
    combined_scores[adj] = f1_score(test_Y, clf_out)
    combined_scores_list.append(v)
    
    #chooser
    test_X = combined_train_test_set[adj]['test']['X']
    test_Y = combined_train_test_set[adj]['test']['Y']
    clf = chooser_classifiers[adj]['classifier']
    clf_out = clf.predict(test_X)
    v = f1_score(test_Y, clf_out)
    chooser_scores[adj] = f1_score(test_Y, clf_out)
    chooser_scores_list.append(v)
    
    #mkl
    mkl_scores_list.append(mkl_scores[adj])
```

# Analysis of the scores

#### Note that white spaces in the bar chart mean that the corresponding classifier has a 0 score (apart from the margin between them)

```python
numitems = len(utilities.adjectives)
ngroups = 5
ind = arange(numitems)
margin = 0.15
width = (1.-2.*margin)/5

xdata = ind+margin+(0*width)
barh(xdata, static_scores_list, width, label="chooser", color="red")

xdata = ind+margin+(1*width)
barh(xdata, dynamic_scores_list, width, label="dynamic", color="green")

xdata = ind+margin+(2*width)
barh(xdata, combined_scores_list, width, label="combined", color="orange")

xdata = ind+margin+(3*width)
barh(xdata, static_scores_list, width, label="static", color="blue")

xdata = ind+margin+(4*width)
barh(xdata, mkl_scores_list, width, label="mkl", color="yellow")

yticks(ind+width*2, utilities.adjectives, rotation=45);
legend();
```

## Who's Better (per Adjective)?

```python
classifiers = ['static', 'dynamic', 'chooser', 'combined', 'mkl']
for adj in utilities.adjectives:
    all_vals = (static_scores[adj], dynamic_scores[adj], chooser_scores[adj], combined_scores[adj], mkl_scores[adj])
    best_idx = argmax(all_vals)
    best_val = all_vals[best_idx]
    avg = mean(all_vals)
    clf = classifiers[best_idx]
    print "Adjective %s, best classifier is %s with a score of %.2f beating a mean score of %.2f" %(adj, clf, best_val, avg)
```
