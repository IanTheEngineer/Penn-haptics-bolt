## Collecting Data, Scroll Below to See the Results

```python
import sklearn
import load_static_features
import utilities
from collections import defaultdict
import cPickle
import os
import train_combined_features
import train_adjectives_features
from sklearn.metrics import f1_score
%pylab inline
figsize(15,5)
```

#### Loading Static Features Classifiers

```python
path = "/home/pezzotto/log/static_features/trained_adjectives"
static_classifiers = {}
for adj in utilities.adjectives:
    filename = "trained_" + adj + ".pkl"
    f = cPickle.load(open(os.path.join(path, filename))) 
    static_classifiers[adj]= f
```

#### Loading Combined Features Classifiers

```python
path = "/home/pezzotto/log/combined_features/trained_adjectives"
combined_classifiers = {}
for adj in utilities.adjectives:
    filename = "trained_" + adj + ".pkl"
    f = cPickle.load(open(os.path.join(path, filename))) 
    combined_classifiers[adj]= f
```

#### Loading Dynamic Features Classifiers

```python
path = "/home/pezzotto/log/hmm_features/trained_adjectives"
dynamic_classifiers = {}
for adj in utilities.adjectives:
    filename = "trained_" + adj + ".pkl"
    f = cPickle.load(open(os.path.join(path, filename))) 
    dynamic_classifiers[adj]= f
```

#### Loading Training and Test Sets

```python
def load_train_test_sets(path):
    train_test_set = defaultdict(dict)    
    static_all_features = train_adjectives_features.load_adjective_phase(path)
    
    for adj in utilities.adjectives:
        train_X, train_Y = train_combined_features.collapse_all_phases(static_all_features, adj, utilities.phases, "train")
        test_X, test_Y = train_combined_features.collapse_all_phases(static_all_features, adj, utilities.phases, "test")
        train_test_set[adj]['train'] = (train_X, train_Y)
        train_test_set[adj]['test'] = (test_X, test_Y)
    
    return train_test_set

#static_train_test_set = load_train_test_sets("/home/pezzotto/log/static_features")
#dynamic_train_test_set = load_train_test_sets("/home/pezzotto/log/hmm_features")
static_train_test_set = train_adjectives_features.load_adjective_collective("/home/pezzotto/log/static_features")
dynamic_train_test_set = train_adjectives_features.load_adjective_collective("/home/pezzotto/log/hmm_features")

#combining the twos for the combined classifier
def combine_datasets(static_set, dynamic_set):
    combined_train_test_set = defaultdict(dict)
    
    for adj in utilities.adjectives:
        static_X = static_set[adj]['train']['X']
        train_Y = static_set[adj]['train']['Y']
        dyn_X = dynamic_set[adj]['train']['X']
        combined_train_test_set[adj]['train'] = (hstack((static_X, dyn_X)), train_Y)
        
        static_X = static_set[adj]['test']['X']
        train_Y = static_set[adj]['test']['Y']
        dyn_X = dynamic_set[adj]['test']['X']
        combined_train_test_set[adj]['test'] = (hstack((static_X, dyn_X)), train_Y)
    
    return combined_train_test_set
combined_train_test_set = combine_datasets(static_train_test_set, dynamic_train_test_set)
```

#### Getting the Results for Static Classifiers

```python
static_test_score = {}
static_train_score = {}
for adj in utilities.adjectives:
    clf_dict = static_classifiers[adj]
    clf = clf_dict['classifier']
    scaler = clf_dict['scaler']
    test_X = static_train_test_set[adj]['test']['X']
    test_Y = static_train_test_set[adj]['test']['Y']
    if scaler:
        clf_out = clf.predict(scaler.transform(test_X))
    else:
        clf_out = clf.predict(test_X)
    static_test_score[adj] = f1_score(test_Y, clf_out)
    
    train_X = static_train_test_set[adj]['train']['X']
    train_Y = static_train_test_set[adj]['train']['Y']
    if scaler:
        clf_out = clf.predict(scaler.transform(train_X))
    else:
        clf_out = clf.predict(train_X)
    static_train_score[adj] = f1_score(train_Y, clf_out)
    
```

#### Getting the Results for Dynamic Classifiers

```python
dynamic_test_score = {}
dynamic_train_score = {}
for adj in utilities.adjectives:
    clf_dict = dynamic_classifiers[adj]
    clf = clf_dict['classifier']
    scaler = clf_dict['scaler']
    test_X = dynamic_train_test_set[adj]['test']['X']
    test_Y = dynamic_train_test_set[adj]['test']['Y']
    if scaler:
        clf_out = clf.predict(scaler.transform(test_X))
    else:
        clf_out = clf.predict(test_X)
    dynamic_test_score[adj] = f1_score(test_Y, clf_out)
    
    train_X = dynamic_train_test_set[adj]['train']['X']
    train_Y = dynamic_train_test_set[adj]['train']['Y']
    if scaler:
        clf_out = clf.predict(scaler.transform(train_X))
    else:
        clf_out = clf.predict(train_X)
    dynamic_train_score[adj] = f1_score(train_Y, clf_out)
```

#### Getting the Results for Combined Classifiers

```python
combined_test_score = {}
combined_train_score = {}
for adj in utilities.adjectives:
    clf_dict = combined_classifiers[adj]
    clf = clf_dict['classifier']
    scaler = clf_dict['scaler']
    test_X = combined_train_test_set[adj]['test'][0]
    test_Y = combined_train_test_set[adj]['test'][1]
    if scaler:
        clf_out = clf.predict(scaler.transform(test_X))
    else:
        clf_out = clf.predict(test_X)
    combined_test_score[adj] = f1_score(test_Y, clf_out)
    
    train_X = combined_train_test_set[adj]['train'][0]
    train_Y = combined_train_test_set[adj]['train'][1]
    if scaler:
        clf_out = clf.predict(scaler.transform(train_X))
    else:
        clf_out = clf.predict(train_X)
    combined_train_score[adj] = f1_score(train_Y, clf_out)
```

# Analysis of the Data

## Static vs Dynamic

```python
plot(static_test_score.values(), '-ro', ms=20, lw=0.5, alpha=0.7, mfc='red', label="static");
plot(dynamic_test_score.values(), '-bo', ms=20, lw=0.5, alpha=0.7, mfc='blue', label="dynamic");
xticks(range(len(static_test_score)), static_test_score.keys(), rotation=45);
legend(loc = "lower right");
mean_static_score = mean(static_test_score.values())
mean_dynamic_score = mean(dynamic_test_score.values())
title("Static vs Dynamic. Mean Static Score: %.2f, Mean Dynamic Score: %.2f" %(mean_static_score, mean_dynamic_score) );
```

## Which One is Better?

```python
diff_scores = {}
for adj in utilities.adjectives:
    #sscore = static_test_score[adj]
    #dscore = dynamic_test_score[adj]
    sscore = static_train_score[adj]
    dscore = dynamic_train_score[adj]
    diff_scores[adj] = sscore-dscore
    
plot(diff_scores.values(), '-o', ms=20, lw=0.5, alpha=0.7, mfc='orange');
plot(range(len(diff_scores)), [0]*(len(diff_scores)), 'r', lw=2)
xticks(range(len(diff_scores)), diff_scores.keys(), rotation=45);
mean_diff_score = mean(diff_scores.values())
title("Positive indicates a better static score, negative is better for dynamics. Mean diff score: %.2f"%mean_diff_score);
```

No approach seems to be dominant, and in many cases they equal each other (zero difference values, along the red line). A few exceptions arise though, e.g. *thick* is classified only by static features, while  *thin*, *crinkly* and *absorbent* work better with dynamic features.

## What if We Take the Best of the Twos?

```python
bests = {}
for adj in utilities.adjectives:
    #sscore = static_test_score[adj]
    #dscore = dynamic_test_score[adj]
    #bests[adj] = sscore if sscore >= dscore else dscore
    sscore = static_train_score[adj]
    dscore = dynamic_train_score[adj]
    bests[adj] = static_test_score[adj] if sscore >= dscore else dynamic_test_score[adj]
    
mean_score = mean(bests.values())

plot(bests.values(), '-o', ms=20, lw=2, alpha=0.7, mfc='orange');
xticks(range(len(bests)), bests.keys(), rotation=45);
title("Mean F1 score: %.2f" % mean_score);
```

## Score for Combined Features

```python
plot(combined_test_score.values(), '-o', ms=20, lw=2, alpha=0.7, mfc='orange');
xticks(range(len(combined_test_score)), combined_test_score.keys(), rotation=45);
mean_score =  mean(combined_test_score.values())
title("Mean F1 score: %.2f" % mean_score);
```

## Is There any Improvement Over Taking the Best of the Twos?

```python
all_stars = {}
for adj in utilities.adjectives:
    if combined_test_score[adj] > bests[adj]:
        print "Adjective %s, combining works better, score is: %.2f against %.2f" %(adj, combined_test_score[adj], bests[adj])
        all_stars[adj] = combined_test_score[adj]
    else:
        all_stars[adj] = bests[adj]
```

## Taking the best of the three worlds then..

```python
plot(all_stars.values(), '-o', ms=20, lw=2, alpha=0.7, mfc='orange');
xticks(range(len(all_stars)), all_stars.keys(), rotation=45);
mean_score =  mean(all_stars.values())
title("Mean F1 score: %.2f" % mean_score);
```

## Robots vs Humans - No Correlation!

```python
#aligning the data per adjective
human_scores = []
robot_scores = []
for adj in utilities.adjectives:
    human_scores.append(utilities.human_scores[adj])
    robot_scores.append(all_stars[adj])
plot(human_scores, '-ro', ms=20, lw=0.5, alpha=0.7, mfc='red', label="human");
plot(robot_scores, '-bo', ms=20, lw=0.5, alpha=0.7, mfc='blue', label="robot");
xticks(range(len(human_scores)), utilities.adjectives, rotation=45);
legend(loc = "lower right");
mean_human_score = mean(human_scores)
mean_robot_score = mean(robot_scores)
title("Robot vs Human. Mean Human Score: %.2f, Mean Robot Score: %.2f" %(mean_human_score, mean_robot_score) );
```
